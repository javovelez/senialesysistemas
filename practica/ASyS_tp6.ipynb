```python
DT=0.001 # time step
RNG=np.array([-5,9]) # time interval s

drng=RNG[1]-RNG[0]
n=np.arange(drng/DT)*DT+RNG[0] #time
h=np.e**(-n)*np.heaviside(n,1)*np.heaviside(5-n,1)*3
x=n*np.heaviside(n,1)*np.heaviside(3-n,1)
y=np.convolve(x,h)

plt.plot(n,h,'r',label=r'h($\tau$)')
plt.plot(n,x,'b',label=r'x($\tau$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$\tau$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_1.png')
plt.show()

DISP=-1
#xt=(DISP-n)*np.heaviside(DISP-n,1)*np.heaviside(3-DISP+n,1)
ht=h=np.e**(-(DISP-n))*np.heaviside((DISP-n),1)*np.heaviside(5-(DISP-n),1)*3

plt.plot(n,ht,'r',label=r'h(t-$\tau$) ; t<0')
plt.plot(n,x,'b',label=r'x($\tau$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$\tau$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_2.png')
plt.show()

DISP=1
#xt=(DISP-n)*np.heaviside(DISP-n,1)*np.heaviside(3-DISP+n,1)
ht=h=np.e**(-(DISP-n))*np.heaviside((DISP-n),1)*np.heaviside(5-(DISP-n),1)*3

plt.fill_between(n, 0, ((ht>0)&(x>0))*3,color='k',alpha=0.2)
plt.plot(n,ht,'r',label=r'h(t-$\tau$) ; 0<t<3')
plt.plot(n,x,'b',label=r'x($\tau$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$\tau$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_3.png')
plt.show()


DISP=4
#xt=(DISP-n)*np.heaviside(DISP-n,1)*np.heaviside(3-DISP+n,1)
ht=h=np.e**(-(DISP-n))*np.heaviside((DISP-n),1)*np.heaviside(5-(DISP-n),1)*3

plt.fill_between(n, 0, ((ht>0)&(x>0))*3,color='k',alpha=0.2)
plt.plot(n,ht,'r',label=r'h(t-$\tau$) ; 3<t<5')
plt.plot(n,x,'b',label=r'x($\tau$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$\tau$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_4.png')
plt.show()



DISP=6
#xt=(DISP-n)*np.heaviside(DISP-n,1)*np.heaviside(3-DISP+n,1)
ht=h=np.e**(-(DISP-n))*np.heaviside((DISP-n),1)*np.heaviside(5-(DISP-n),1)*3

plt.fill_between(n, 0, ((ht>0)&(x>0))*3,color='k',alpha=0.2)
plt.plot(n,ht,'r',label=r'h(t-$\tau$) ; 5<t<8')
plt.plot(n,x,'b',label=r'x($\tau$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$\tau$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_5.png')
plt.show()


DISP=9
#xt=(DISP-n)*np.heaviside(DISP-n,1)*np.heaviside(3-DISP+n,1)
ht=h=np.e**(-(DISP-n))*np.heaviside((DISP-n),1)*np.heaviside(5-(DISP-n),1)*3

plt.fill_between(n, 0, ((ht>0)&(x>0))*3,color='k',alpha=0.2)
plt.plot(n,ht,'r',label=r'h(t-$\tau$) ; t>8')
plt.plot(n,x,'b',label=r'x($\tau$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$\tau$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_6.png')
plt.show()

plt.plot(np.arange(len(y))*DT+2*RNG[0],y*DT,'k',label=r'y($t$)')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel(r'$t$')
plt.legend()
plt.xlim(RNG)
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp6_7_7.png')
plt.show()
```
