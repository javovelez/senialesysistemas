Ej. 1.a)

```python
n=np.arange(10)+1
zn=n+n*1j
for x in n-1:
    zn[x]=1j*n[x]*(n[x]+1)
X = np.real(zn)
Y = np.imag(zn)
plt.plot(X,Y, '-ob')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('x=Re(z)')
plt.ylabel('y=Im(z)')
plt.axis('equal')
#plt.xlim([-3,3])
#plt.ylim([-4,5])
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp4_1a.png')
plt.show()
```

Ej. 1.b)

```python
n=np.arange(100)+1
zn=n+n*1j
for x in n-1:
    zn[x]=(1+2j)**n[x]/np.math.factorial(n[x])
X = np.real(zn)
Y = np.imag(zn)
plt.plot(X,Y, '-ob')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('x=Re(z)')
plt.ylabel('y=Im(z)')
plt.axis('equal')
#plt.xlim([-3,3])
#plt.ylim([-4,5])
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp4_1b.png')
plt.show()
```

Ej. 1.c)

```python
n=np.arange(100)+1
zn=n+n*1j
for x in n-1:
    zn[x]=(1j)**n[x]*np.cos(n[x]*np.pi)
X = np.real(zn)
Y = np.imag(zn)
plt.plot(X,Y, '-ob')
plt.grid('true')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel('x=Re(z)')
plt.ylabel('y=Im(z)')
plt.axis('equal')
#plt.xlim([-3,3])
#plt.ylim([-4,5])
if PLOT == 1:
    plt.savefig(OUTPUT_PATH+'/tp4_1c.png')
plt.show()
```
